Import Pandas

```python
import pandas as pd
```

Read the `pokemon.csv` file and set the index to `['Type 1', 'Name']`

```python
pokemon = pd.read_csv('pokemon.csv', index_col=['Type 1', 'Name'])
pokemon.head()
```

Sort the pokemons by `Type 1`

```python
pokemon.sort_values('Type 1', inplace=True)
pokemon
```

Select all the `Fire` pokemons that start with the letter `C`

```python
fire_pokemon = pokemon.loc['Fire']
#starts_with_c = pokemon.index.str.startswith('C')

#fire_pokemon[starts_with_c]
```

Using `agg()` get the `max`, `min` and the `mean` of all the pokemons for the columns `['Attack', 'Defense']` and round them to 1 decimal place.

```python
pokemon.agg({'Attack': ['max','min','mean'], 'Defense': ['max','min','mean']})
```

Create a function that checks the Attack of a pokemon. If the `Attack` is over 100, it should now display `'STRONG'`, if it is less than that, it should display `'WEAK'`. Update the DataFrame to display this new Attack column`

```python
def strength(attack: int) -> str:
    if attack > 100:
        return 'STRONG'
    else:
        return 'WEAK'
```

```python
pokemon['Attack'] = pokemon['Attack'].apply(strength)
pokemon
```

Create a new column called `Reversed` which holds each pokemon name reversed

```python
reversed_name = pokemon.index.get_level_values('Name').str[::-1]
pokemon['Reversed'] = reversed_name
pokemon
```
